it is okay that it is not exacly linear, but we can see a tendency. we can relate the p-ratio to the norm_sp though a linear or higher degree equation and use this 
as the way to get the p-ratio for 2025, 2035 and 2045. (instead of using the historical values from 2021)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

start_year_indexes = [0, 8785, 17545, 26305, 35065, 43849, 52608, 61368]
year_labels = [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

pr_0_6 = pd.read_excel('p_ratios_0_6.xlsx')
# m_NL_0_6	m_DK2_0_6	m_BE_0_6	m_FR_0_6	m_FIN_0_6	m_PL_0_6
pr_0_7= pd.read_excel('p_ratios_0_7.xlsx')

sp_norm_0_6= pd.read_excel('norm_sp_0_6.xlsx')
sp_norm_0_7= pd.read_excel('norm_sp_0_7.xlsx')

# SP_FR	    SP_BE	    SP_DE	    SP_PL	    SP_FIN	    SP_NL	    SP_DK2

# scatterplot
x = pr_0_7['m_FR_0_7'].iloc[0:52608]
y = sp_norm_0_7['SP_FR'].iloc[0:52608]

nan_mask = np.isnan(x) | np.isnan(y)

x_no_nan = x[~nan_mask]
y_no_nan = y[~nan_mask]

m, b = np.polyfit(x_no_nan, y_no_nan, deg=1)

line = f'y = {m:.2f}x + {b:.2f}'

plt.scatter(x, y, s=0.2, color='lightgreen')
plt.plot(x, m * x + b, color='red', label=line)
plt.xlabel('P Ratios')
plt.ylabel('SP Norm')
plt.title('Scatter Plot FR: p-ratio vs sp_norm between 2016-2021')
plt.legend()
plt.show()

```

```python
selected_x = 1  
selected_y = m * selected_x + b

print(f"For x = {selected_x:.2f}, the corresponding y value is {selected_y:.2f}")
```
