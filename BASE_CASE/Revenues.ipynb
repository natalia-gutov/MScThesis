```python
# Capacity Factors
#  by definition: The ratio of the electrical energy produced by a generating unit for the period of time considered to the electrical energy that could have been produced at continuous full power operation during the same period.
# CF P_100
# CF_100_1_data = (P_100_1_data.sum() / (cap_inst * 8736)) * 100
# CF_100_1 = f'{CF_100_1_data.item():.2f}%'
locs_w = ['LOC_1', 'LOC_2', 'LOC_3', 'LOC_4.1', 'LOC_5.1', 'LOC_6', 'LOC_7', 'LOC_8.1', 'LOC_9', 'LOC_10']
CF_100 = {}
CF_150 = {}
for loc in locs_w:
    P_100data = df_P_100_year[loc][24:-24]*cap_inst
    # CF_100data = (P_100data.sum() / (cap_inst * 8736))
    CF_100data = (P_100data.sum() / (cap_inst * 8736))
    CF100 = f'{CF_100data.item() * 100:.2f}%'
    CF_100[loc] = CF100
    P_150data = df_P_150_year[loc][24:-24]*cap_inst
    CF_150data = (P_150data.sum() / (cap_inst * 8736))
    CF150 = f'{CF_150data.item() * 100:.2f}%'
    CF_150[loc] = CF150

CF_PV = {}
for loc in locs_s:
    P_s_data = df_P_PV_year[loc][24:-24]*cap_inst
    CF_PVdata = (P_s_data.sum() / (cap_inst * 8736))
    CFPV = f'{CF_PVdata.item()*100:.2f}%'
    CF_PV[loc] = CFPV
```

```python
# Definir las hojas de Excel y las columnas a extraer
sheets_to_read = ['2025', '2035', '2045']

columns_to_extract = ['Time (UTC)', 'BE', 'NL', 'FR', 'PL', 'FIN', 'DK2', 'DE4-S', 'DE4-N', 'DE4-W']

# Nombre de los archivos de Excel
file_names = ['R1.xlsx', 'R4.xlsx', 'R19.xlsx']

# Diccionario para almacenar los DataFrames
dataframes = {}

# Leer los archivos de Excel y guardar los datos en DataFrames
for file_name in file_names:
    # Leer el archivo de Excel
    df = pd.read_excel(file_name, sheet_name=None)
    
    # Crear un diccionario para almacenar los DataFrames por hoja
    df_dict = {}
    
    # Leer las hojas de Excel y extraer las columnas necesarias
    for sheet_name in sheets_to_read:
        df_sheet = df[sheet_name][columns_to_extract]
        df_dict[sheet_name] = df_sheet
    
    # Agregar el diccionario de DataFrames al diccionario principal
    dataframes[file_name] = df_dict

# Acceder a los DataFrames según el archivo y la hoja
for file_name, df_dict in dataframes.items():
    for sheet_name, df_sheet in df_dict.items():
        df_name = f"df_{file_name.split('.')[0]}_{sheet_name}"
        globals()[df_name] = df_sheet  # Guardar el DataFrame en una variable con nombre dinámico

# Imprimir los nombres de los DataFrames creados
print("DataFrames creados:")
for file_name in file_names:
    for sheet_name in sheets_to_read:
        df_name = f"df_{file_name.split('.')[0]}_{sheet_name}"
        df = globals()[df_name]  # Obtener el DataFrame
        
        plt.plot(df['Time (UTC)'], df['BE'], label='BE')
        plt.plot(df['Time (UTC)'], df['NL'], label='NL')
        plt.plot(df['Time (UTC)'], df['FR'], label='FR')
        plt.plot(df['Time (UTC)'], df['PL'], label='PL')
        plt.plot(df['Time (UTC)'], df['FIN'], label='FIN')
        plt.plot(df['Time (UTC)'], df['DK2'], label='DK2')
        plt.plot(df['Time (UTC)'], df['DE4-S'], label='DE4-S')
        plt.plot(df['Time (UTC)'], df['DE4-N'], label='DE4-N')
        plt.plot(df['Time (UTC)'], df['DE4-W'], label='DE4-W')
        
        plt.xlabel('Time (UTC)')
        plt.ylabel('Eur/Mwh')
        plt.title(f'Spot Prices for {df_name}')
        plt.legend()
        plt.show()
```

```python
import os
# rev for wind technologies
tech = ['100', '150']
CorRES_loc = ['1', '2', '3', '4_1', '5_1', '6', '7', '8_1', '9', '10']
SP_loc = ['DE4-S', 'DE4-N', 'PL', 'FIN', 'NL', 'BE', 'DK2']
R_sc = ['R1', 'R4', 'R19']
year_sc = ['2025', '2035', '2045']

revenues = {} 

for t in tech:
    for cl in CorRES_loc:
        for rs in R_sc:
            for ys in year_sc:
                # Check the specific constraints for each CorRES_loc value
                if cl == '1':
                    sp_loc = 'DE4-S'
                elif cl == '2':
                    sp_loc = 'DE4-N'
                elif cl == '3':
                    sp_loc = 'PL'
                elif cl == '4_1':
                    sp_loc = 'FIN'
                elif cl == '5_1':
                    sp_loc = 'NL'
                elif cl == '6':
                    sp_loc = 'BE'
                elif cl in ['7', '8_1']:
                    sp_loc = 'DK2'
                elif cl == '9':
                    sp_loc = 'DE4-S'
                elif cl == '10':
                    sp_loc = 'PL'
                else:
                    sp_loc = ''

                # format
                rev_var = f'rev_{t}_{cl}_{rs}_{ys}'

                # Checkk
                if rev_var not in revenues:
                    # Assuming scenario 2025
                    P_data = globals()[f'P_{t}_{cl}_data']
                    df_data = globals()[f'df_{rs}_{ys}']

                    P_data = P_data.reset_index(drop=True)
                    df_data = df_data.reset_index(drop=True)
                    df_data = df_data.reindex(P_data.index)  # Reindex df_data using P_data index
                    df_data[sp_loc] = df_data[sp_loc].reset_index(drop=True)

                    # Check if matiching index
                    if P_data.index.equals(df_data.index):
                        # Calculate the revenue
                        revenue = P_data[f'LOC_{cl}'] * df_data[sp_loc]
                        # Store in the dictionary
                        revenues[rev_var] = revenue
                    else:
                        print("Error: Indices do not match for", rev_var)

# Print the calculated revenues
# for rev_var, revenue in revenues.items():
#     print(f'{rev_var}:')
#     print(revenue)
#     print()
rev_var = 'rev_100_10_R19_2045'  
revenue_value = revenues.get(rev_var)

if revenue_value is not None:
    cumulative_sum = revenue_value.cumsum()
    tot_ann_rev = cumulative_sum[-1:]
    # plt.plot(revenue_value.index, revenue_value)
    # plt.plot(cumulative_sum.index, cumulative_sum, highest_value,'ro', label= f'tot yearly revenue={highest_value}')
    # plt.plot(cumulative_sum.index, cumulative_sum, label='Cumulative Hourly Revenue')
    # plt.plot(len(cumulative_sum) - 1, tot_ann_rev, 'ro', label=f'tot yearly revenue={tot_ann_rev.values}')
    # plt.legend()
    # plt.xlabel('Time')
    # plt.ylabel('Cumulative Revenue')
    # plt.title(f'Cumulative Revenue with for {rev_var}')
    # plt.xticks(np.arange(0, len(cumulative_sum), 1000))
    # plt.grid(True)
    # plt.show()
    print(f'{rev_var}', tot_ann_rev.values)
    # empty photo:
    # filename = f'{rev_var}.png'  # Specify the filename
    # save_path = os.path.join(r'C:\Users\ngt13\Desktop\Thesis_main_folder\BASE_CASE\Graphs_py\Revenues', filename)
    # plt.savefig(save_path)
        
else:
    print(f"{rev_var} does not exist in the revenue dictionary.")
```

```python
# rev for solar:
tech = ['PV']
CorRES_loc = ['1', '2', '3', '4_2', '5_2', '6', '7', '8_2', '9', '10']
SP_loc = ['DE4-S', 'DE4-N', 'PL', 'DE4-W','FR', 'BE', 'DK2']
R_sc = ['R1', 'R4', 'R19']
year_sc = ['2025', '2035', '2045']

revenues_s = {}  # Dictionary to store the revenue series

for t in tech:
    for cl in CorRES_loc:
        for rs in R_sc:
            for ys in year_sc:
                # Check the specific constraints for each CorRES_loc value
                if cl == '1':
                    sp_loc = 'DE4-S'
                elif cl == '2':
                    sp_loc = 'DE4-N'
                elif cl == '3':
                    sp_loc = 'PL'
                elif cl == '4_2':
                    sp_loc = 'DE4-W'
                elif cl == '5_2':
                    sp_loc = 'FR'
                elif cl == '6':
                    sp_loc = 'BE'
                elif cl == '7':
                    sp_loc = 'DK2'
                elif cl == '8_2':
                    sp_loc = 'FR'
                elif cl == '9':
                    sp_loc = 'DE4-S'
                elif cl == '10':
                    sp_loc = 'PL'
                else:
                    sp_loc = ''

                # Construct the variable name
                rev_var_s = f'rev_{t}_{cl}_{rs}_{ys}'

                # Check if the revenue series is already calculated
                if rev_var_s not in revenues_s:
                    # Assuming scenario 2025
                    P_data = globals()[f'P_{t}_{cl}_data']
                    df_data = globals()[f'df_{rs}_{ys}']

                    P_data = P_data.reset_index(drop=True)
                    df_data = df_data.reset_index(drop=True)
                    df_data = df_data.reindex(P_data.index)  # Reindex df_data using P_data index
                    df_data[sp_loc] = df_data[sp_loc].reset_index(drop=True)

                    # Check if the indices match
                    if P_data.index.equals(df_data.index):
                        # Calculate the revenue
                        revenue = P_data[f'LOC_{cl}'] * df_data[sp_loc]

                        # Store the revenue series in the dictionary
                        revenues_s[rev_var_s] = revenue
                    else:
                        print("Error: Indices do not match for", rev_var_s)

# Print the calculated revenues
# for rev_var, revenue in revenues.items():
#     print(f'{rev_var}:')
#     print(revenue)
#     print()
rev_var_s = 'rev_PV_1_R1_2025'  # Replace with the actual variable name you want to retrieve
revenue_value_s = revenues_s.get(rev_var_s)

if revenue_value_s is not None:
    cumulative_sum_s = revenue_value_s.cumsum()
    highest_value_s = cumulative_sum_s[-1:]
    # plt.plot(revenue_value.index, revenue_value)
    # plt.plot(cumulative_sum.index, cumulative_sum, highest_value,'ro', label= f'tot yearly revenue={highest_value}')
    plt.plot(cumulative_sum_s.index, cumulative_sum_s, label='Cumulative Hourly Revenue')
    plt.plot(len(cumulative_sum_s) - 1, highest_value_s, 'ro', label=f'tot yearly revenue={highest_value_s.values}')
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('Cumulative Revenue')
    plt.title(f'Cumulative Revenue with for {rev_var_s}')
    plt.xticks(np.arange(0, len(cumulative_sum_s), 1000))
    plt.grid(True)
    plt.show()
        
else:
    print(f"{rev_var_s} does not exist in the revenue dictionary.")
```
